```python
import pandas as pd
import os


folder_path_input = "/Users/elboyer/Desktop/openlca_dataset/inputs"
folder_path_output = "/Users/elboyer/Desktop/openlca_dataset/outputs"


results = []

for file_name in os.listdir(folder_path_input):
    if file_name.endswith(".xlsx") or file_name.endswith(".xls"):
        file_path = os.path.join(folder_path_input, file_name)
        try:
            df_input = pd.read_excel(file_path, sheet_name="Inputs")

            # Filter rows
            nitrogen_rows = df_input[df_input['Flow'] == "inorganic nitrogen fertiliser, as N"]
            phosphorus_rows = df_input[df_input['Flow'] == "inorganic phosphorus fertiliser, as P2O5"]

     
            avg_amount_nitrogen = nitrogen_rows['Amount'].mean()
            avg_amount_phosphorus = phosphorus_rows['Amount'].mean()

            
            product_name = file_name.split()[0].split('_')[0]


            results.append({
                "product": product_name,
                "input_nitrogen per kg of product": avg_amount_nitrogen,
                "input_phosphorus per kg of product": avg_amount_phosphorus
            })

        except Exception as e:
            print(f"Error in input file gsg{file_name}: {e}")


for file_name in os.listdir(folder_path_output):
    
    if file_name.endswith(".xlsx") or file_name.endswith(".xls"):
        file_path = os.path.join(folder_path_output, file_name)
        try:
            df_output = pd.read_excel(file_path, sheet_name="Impacts")

            # Set header row and clean it up
            df_output.columns = df_output.iloc[0]
            df_output = df_output[1:]

            # Find eutrophication rows
            fresh_row = df_output[df_output["Impact category"] == "freshwater eutrophication - FEP"]
            marine_row = df_output[df_output["Impact category"] == "marine eutrophication - MEP"]

            
            fresh_val = fresh_row['Result'].values[0] if not fresh_row.empty else None
            fresh_unit = fresh_row['Reference unit'].values[0] if not fresh_row.empty else None
            marine_val = marine_row['Result'].values[0] if not marine_row.empty else None
            marine_unit = marine_row['Reference unit'].values[0] if not marine_row.empty else None

            # Print for debugging
            print(file_name)
            product_name = file_name.split()[0].split('_')[0]
            print(product_name)
     

            for item in results:
                if item["product"] == product_name:
                    item["freshwater_eutrophication (kg P-Eq)"] = fresh_val
                    item["unit_freshwater"] = fresh_unit
                    item["marine_eutrophication (kg N-Eq)"] = marine_val
                    item["unit_marine"] = marine_unit
                    break

        except Exception as e:
            print(f"Error in output file {file_name}: {e}")


df_final = pd.DataFrame(results)
df_final


output_file_path = "./final_results.xlsx"
df_final.to_excel(output_file_path, index=False)
```
