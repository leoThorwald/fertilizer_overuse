```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Swiss_data.csv')
data.head()
```

```python
print('data shape:', data.shape)
print('data columns:', data.columns)
print('data unique:\n', data.nunique())
 
data.drop(labels = ['Domain Code', 'Element Code', 'Domain', 'Area Code (M49)', 'Item Code', 'Year Code', 'Item'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
# Pivot the table to have years as rows and elements as columns
df_pivot = data.pivot(index="Year", columns="Element", values=["Unit", "Value"])
df_pivot.columns = [f"{col[1]}_{col[0]}" for col in df_pivot.columns]
df_pivot.reset_index(inplace=True)
df_pivot.rename(columns={"Year": "Date"}, inplace=True)

df_pivot.isnull().sum()
```

```python

df_pivot[df_pivot['Date'] > 1999][['Date', 'Production_Value']].rename(columns = {'Production_Value': 'Swiss_production'})
```
