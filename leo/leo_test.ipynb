```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("fertilizers.csv")
```

```python
df2 = df.loc[df["Unit"] == "kg/ha", ["Year", "Value", "Area", "Item"]]
df2.head()
```

```python
df2.groupby("Item")["Value"].mean().plot(kind="bar")
```

```python
cross = pd.crosstab(df2["Area"], df2["Item"], values=df2["Value"], aggfunc="mean")
sns.heatmap(cross, cmap="coolwarm", annot=True)
```
